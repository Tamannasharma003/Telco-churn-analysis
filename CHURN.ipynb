```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/Users/tushar/Downloads/Telco-Customer-Churn.csv")
```

```python
#replaced blanks space with 0 as tenure is 0 so no charges are recorded 
```

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ",0)
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

```python
#senior citizen value 0 and 1 changed to yes and no
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python
df.head()
```

```python
# seeing churn data using sns
```

```python
a=sns.countplot(data=df,x="Churn")
a.bar_label(a.containers[0])
sns.set_style("darkgrid")
plt.title("COUNT OF CHURN",fontsize=20)
plt.show()
```

```python
#FROM THIS COUNTPLOT WE SEE 5174 PEOPLE CHURNED OUT OF 7043
```

```python
#in pie chart
Group=df.groupby("Churn").agg({"Churn":"count"})
ex=[0,0.2]
plt.figure(figsize=(4,4))
plt.pie(Group["Churn"],labels=Group.index,shadow=True,autopct="%.2f",explode=ex,colors=["pink","green"])
plt.title("PERCENTAGE COUNT OF CHURN",fontsize=15)
plt.show()
```

```python
#FROM THIS PIE CHART WE SEE HOW MANY PERCENTAGED PEOPLE CHURNED OUT
```

```python
ab=sns.countplot(data=df,x="gender",hue="Churn",palette="Set1")
for container in ab.containers:
    ab.bar_label(container)
plt.title("Churn by gender",fontsize=15)
plt.show()
```

```python
#from this countplot we see how many people churned out on the basis of gender
```

```python
ab = sns.countplot(data=df, x="SeniorCitizen", hue="Churn",palette=["yellow","green"])
for container in ab.containers:
    ab.bar_label(container,fontsize=10)
plt.title("Churn by SENIOR CITIZEN", fontsize=20)
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.show()
```

```python
#compartive a greater no of senior citizens have churned
```

```python
plt.figure(figsize=(9,4))
sns.histplot(data=df,x="tenure",hue="Churn", bins = 50)
plt.show()
```

```python
#FROM THIS HISTPLOT WE SEE HOW MANY PEOPLE CHURNED OUT ON THE BASIS OF TENURE
```

```python
ab=sns.countplot(data=df,x="Contract",hue="Churn")
for container in ab.containers:
    ab.bar_label(container,fontsize=10)
plt.title("Churned out on the basis of contract")
plt.show()
```

```python
#people who used services for a long time they stayed but new comers churned out
```

```python
df.columns.values
```

```python
cols = ['PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies']
sns.set_style("darkgrid")

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.countplot(data=df,x=col,hue=df["Churn"],palette="Set2",ax=axes[i])
    axes[i].set_title(col, fontsize=15)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
plt.tight_layout()
plt.show()
```

#The majority of customer who do not churn tend to have services like PhoneServices,InternetServices(particularly DSL),and Online Security enabled.For sevices like OnlineBackup,TechSupport and StreamingTV,churn rates are noticiably hiher when these services are not used or are unavailavle

```python
ab=sns.countplot(data=df,x="PaymentMethod",hue="Churn")
for container in ab.containers:
    ab.bar_label(container)
plt.legend(title="Churn",loc="upper right",bbox_to_anchor=(1,1))
plt.xticks(rotation=45)
plt.show()
```
